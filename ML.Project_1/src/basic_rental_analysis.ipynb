### Примеры машинного обучения в реальной жизни ###
1.Умные рекомендации (например, онлайн-кинотеатры (Netflix), онлайн-маркетплейсы (Ozon), книжные интернет-магазины (Читай-город)). Рекомендательные системы анализируют поведение пользователей и предлагают фильмы, сериалы или товары на основе их предпочтений и истории просмотров/покупок.

Преимущества:
+ Пользователи получают рекомендации, которые соответствуют их интересам, что повышает их удовлетворённость приложением и компанией.
+ Рекомендации могут способствовать дополнительным покупкам.


2.Обнаружение мошенничества в банковских транзакциях. Банки используют алгоритмы машинного обучения для анализа транзакций и выявления аномалий, которые могут указывать на мошенничество.

Преимущества:
+ Системы могут почти мгновенно обнаруживать подозрительные действия и предотвращать финансовые потери.
+ Постоянное обучение моделей на новых данных позволяет повышать точность обнаружения мошенничества.


3.Беспилотные автомобили. Они используют машинное обучение для обработки данных с датчиков (камеры, радары) и принятия решений о движении в реальном времени.

Преимущества:
+ Снижение числа аварий за счет автоматизации управления автомобилем и устранения человеческого фактора
+ Освобождение времени для пассажиров, которые могут заниматься другими делами во время поездки.


4.Обработка естественного языка. Например, чат-боты используют алгоритмы машинного обучения для понимания и генерации человеческой речи и могут взаимодействовать с пользователем

Преимущества:
+ Чат-боты могут работать круглосуточно, предоставляя поддержку пользователям в любое время.
+ Автоматизация ответов на часто задаваемые вопросы снижает нагрузку на службу поддержки.


5.Прогнозирование спроса в розничной торговле. Компании используют машинное обучение для анализа данных о продажах и прогнозирования будущего спроса на товары.

Преимущества:
+ Помогает избежать недостатка или избытка товаров на складе, что снижает затраты.
+ Позволяет более точно планировать маркетинговые кампании и акции.


### Классификация задач ###

**Задача 1: Прогноз цены на дом**

Относится к типу "регрессия", т.к. мы предсказываем действительное значение (цену дома) на основе характеристик домов.

**Задача 2: Предсказать, вернет ли клиент кредит**

Относится к типу "классификация", а точнее - "бинарная классификация", т.к. мы предсказываем бинарный результат (вернет/не вернет кредит) на основе данных о клиентах.

**Задача 3: Предсказать, когда пациенту необходимо принять лекарство**

Можно рассматривать как задачу ипа регрессия, если нам надо предсказать конкретное время приёма, или как классификация, если  мы классифицируем время в категории "утро", "вечер", "перед едой", "после еды".

**Задача 4: Выберите, какое лекарство из имеющихся в наличии следует принять пациенту**

Задача типа "классификация" - выбираем один из нескольких классов (лекарств) на основе медицинских данных.

**Задача 5: Выберите сегмент клиентов для промо-коммуникации**

Тип задачи - "кластеризация", группируем клиентов по их поведению для рекламной кампании.

**Задача 6: Распознавание дефектной продукции на производственной линии (на основе фотосканирования)**

Задача типа "классификация", мы распознаём образы и классифицируем продукцию как "дефектную" или "нормальную" на основе изображений.

**Задача 7: Решите, как разместить продукцию на полке в магазине**

Относится к типу "ассоциация", т.к. предложенный опыт прошлого размещения товаров позволяет вывести ассоциативные правила вроде "Люди склонны больше покупать А, если оно лежит рядом с Б".

**Задача 8: Поиск сайтов для ввода текстового запроса**

Относится к типу "классификация", т.к. мы классифицируем сайты по релевантности к запросу пользователя.

**Задача 9: Разделите покупателей магазина на сегменты, чтобы понять различия в их поведении**

Тип задачи - "кластеризация", группируем покупателей по их характеристикам и истории покупок для выявления сегментов.

**Задача 10: Обнаружение аномалий в трафике сайта**

Задача типа "классификация", мы пытаемся классифицировать трафик как "нормальный" или "аномальный" на основе обнаруженных аномалий.

**Пример 1: Умные рекомендации**

Задача типа "ассоциация" и "классификация". Мы используем ассоциативные правила для формирования рекомендаций, а для вывода самих правил классифицируем пользователей и товары. Может также быть отнесена к типу "кластеризация" - если для классификации пользователей не используются заранее известные критерии, например, возраст или ключевые слова в профиле.

**Пример 2: Обнаружение мошенничество в банковских транзакциях**

Задача типа "классификация" - мы определяем, является транзакция мошеннической или нет.

**Пример 3: Беспилотные автомобили**

Задача включает в себя тип "классификация" - для распознавания объектов, и тип "регрессия" - для предсказания движения

**Пример 4: Обработка естественных языков, общение с пользователем**

Включает в себя задачу расшифровки запроса - тип "классификация" и задачу генерации ответа - тип "регрессия".

**Пример 5: Прогнозирование спроса в торговле**

Задача типа "регрессия", по адналогии с ценами на дом. 


### Мультикласс и мультиметка ###

Задачи классификации подразумевают отнесение объекта к тому или иному классу в зависимости от того, обладает ли он характеристикой этого класса (меткой). В зависимости от того, сколько классов могут соответствовать объекту, их можно условно разделить на задачи с несколькими классами (мультиклассовые задачи) и задачи с несколькими метками (мультиметочные задачи).

**Мультиклассовые задачи** - задачи, в которых есть несколько (как правило, более двух, иначе - бинарная классификация) классов, и каждому объекту может соответствовать только один правильный класс. Пример - распознавание животных на фотографиях. Животное может быть кошкой, собакой, человеком, но не несколькими сразу.

**Мультиметочные задачи** - задачи, в которых объект может обладать сразу несколькими характеристиками (метками), соответствующими разным классам, и может быть отнесён к нескольким классам. Пример - распознавание темы текста. Текст может быть одновременно и о программировании, и о тайм-менеджменте. 

**Выходные данные**: в мультиклассовых задачах представляют собой одну метку, указывающую на класс объекта, в мультиметочных - вектор бинарных значений, где каждое значение указывает, принадлежит объект этому классу или нет.

**Метрики**: Ещё одно различие - в метриках. Для мультиклассовых задач используются такие метрики, как точность (accuracy), F1-score, для мультиметочных - Hamming loss или Jaccard index.


### Регрессия и классификация ###

**Является ли пример с ценами на жилье из теории подтипом регрессионной проблемы?** 

Да, этот пример является подтипом регрессионной проблемы. В данном случае мы пытаетемся предсказать непрерывную переменную (цену на жилье) на основе других непрерывных или категориальных переменных (количество спален, ванных комнат и уровень спроса). Это классическая задача регрессии, где цель — предсказать числовое значение.

**Можно ли свести регрессионную проблему к классификации?**

Да, в ряде случаем это можно сделать посредством определённых шагов. Однако нужно учитывать, что при попытке предсказания непрерывной величины, как в примере с ценами, это может повлечь за собой потери в точности, и делать это следует тогда, когда цель - работать с категориями и классами. Например, в задаче с ценами можно работать не с конкретными числами, а с ценовыми категориями (дешевое жильё, средней стоимости, премиум-класса). Также это можно сделать при оценке качества продукта (высокое, среднее, низкое на основании какого-нибудь количественного параметра) или, скажем, прогнозировании времени ожидания в МФЦ (от 5 до 10 минут, от 10 до 15 минут, от 15 и более минут (с нас бесплатный кофе!)).

Есть несколько способов свести задачу регрессии к задаче классификации:


1.Дискретизация целевой переменной:

Диапазон значений переменной можно разбить на категории (об этом уже говорилось выше), тогда регрессионная задача становится классификационной.


2.Классификация по квантилям:

Можно разделить данные на квартильные группы (25%, 50%, 75%) и классифицировать каждую запись в одну из этих групп.


3.Использование методов классификации для предсказания:

Некоторые алгоритмы машинного обучения могут быть адаптированы для работы с задачами регрессии и классификации. Например, деревья решений могут использоваться как для регрессии, так и для классификации (Random Forest).

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
from sklearn.metrics import r2_score
import lightgbm 
import scipy 
import statsmodels 
import matplotlib.pyplot as plt
import seaborn as sns
import json

with open("../datasets/train.json", "r", encoding='utf-8') as file:
    data = json.load(file) 
df = pd.DataFrame(data)
print(df.shape)
print(', '.join(df.columns))
print("'Rent' column is a target")
print(df.info())
print(df.describe())
print(df.corr(numeric_only=True))
train_df = pd.DataFrame({'bathrooms': df['bathrooms'], 'bedrooms': df['bedrooms'], 'interest_level': df['interest_level'], 'price': df['price']})
```

```python
# Target analysis

train_df.hist(column='price', bins=50, edgecolor='black')
plt.show()
train_df.boxplot(column='price', figsize=(10, 5), vert=False)
# sns.boxplot(train_df, x='price')
plt.show()
lower_percentile = train_df['price'].quantile(0.01)
upper_percentile = train_df['price'].quantile(0.99)
train_df = train_df[(train_df['price'] > lower_percentile) & (train_df['price'] < upper_percentile)]
train_df.hist(column='price', bins=50, edgecolor='black')
plt.show()
train_df.boxplot(column='price', figsize=(10, 5), vert=False)
plt.show()
```

```python
# Characteristics Analysis

print(f"The type of column 'interest_level' is {train_df['price'].dtype}")
print(train_df['interest_level'].value_counts())

interest_order = pd.CategoricalDtype(categories=['low', 'medium', 'high'], ordered=True)
train_df['interest_level'] = train_df['interest_level'].astype(interest_order).cat.codes
print(train_df['interest_level'].value_counts())

df['bathrooms'].hist()
plt.show()
df['bathrooms'].plot.box(vert=False, figsize=(10, 5))
plt.show()
df['bedrooms'].hist()
plt.show()
df['bedrooms'].plot.box(vert=False, figsize=(10, 5))
plt.show()
```

```python
# Creating Features

import sklearn.preprocessing

work_df_features = train_df.copy()
work_df_features['bathrooms_squared'] = work_df_features['bathrooms'] * work_df_features['bathrooms']
work_df_features['bedrooms_squared'] = work_df_features['bedrooms'] * work_df_features['bedrooms']
work_df_features['interest_level_squared'] = work_df_features['interest_level'] * work_df_features['interest_level']
work_df_features['room_sum'] = work_df_features['bathrooms'] + work_df_features['bedrooms']
work_df_features['bedrooms_diff'] = work_df_features['bedrooms'].diff()
work_df_features['bathrooms_diff'] = work_df_features['bathrooms'].diff()
work_df_features['interest_level_diff'] = work_df_features['interest_level'].diff()
work_df_features['bed/bath'] = work_df_features['bedrooms'] / work_df_features['bathrooms']
work_df_features['bed-bath'] = work_df_features['bedrooms'] - work_df_features['bathrooms']
work_df_features['bed*bath'] = work_df_features['bedrooms'] * work_df_features['bathrooms']
work_df_features['bedrooms_sqrt'] = work_df_features['bedrooms'] ** 0.5
work_df_features['bathrooms_sqrt'] = work_df_features['bathrooms'] ** 0.5
work_df_features['interest_level_sqrt'] = work_df_features['interest_level'] ** 0.5
work_df_features['2_bedrooms'] = work_df_features['bedrooms'] > 2

print(work_df_features.corr()['price'])
print(work_df_features.corr())

# Applying PolynomialFeatures to train and test data

with open('../datasets/test.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
dataframe = pd.DataFrame(data)

test_df = pd.DataFrame({'bathrooms': dataframe['bathrooms'], 'bedrooms': dataframe['bedrooms'], 'price': dataframe['price']})

# lower_percentile = test_df['price'].quantile(0.01)
# upper_percentile = test_df['price'].quantile(0.99)
# test_df = test_df[(test_df['price'] > lower_percentile) & (test_df['price'] < upper_percentile)]

# train_df = train_df[(train_df['bathrooms'] <= 4)]
# train_df = train_df[(train_df['bedrooms'] <= 6)]
```

```python
train_x_interest = train_df[['bathrooms', 'bedrooms']].copy()
train_y_interest = train_df['interest_level'].copy()

test_x_interest = test_df[['bathrooms', 'bedrooms']].copy()
```

```python
# Training preparation

result_MAE = pd.DataFrame(columns=['model', 'train', 'test'])
result_RMSE = pd.DataFrame(columns=['model', 'train', 'test'])
result_R2 = pd.DataFrame(columns=['model', 'train', 'test'])
```

```python
# Linear Regression
regression_model_interest = LinearRegression()
regression_model_interest.fit(train_x_interest, train_y_interest)
prediction_interest = regression_model_interest.predict(test_x_interest)
test_df['interest_level'] = prediction_interest

train_x_raw = train_df[['bathrooms', 'bedrooms', 'interest_level']].copy()
train_y = train_df['price'].copy()

test_x_raw = test_df[['bathrooms', 'bedrooms', 'interest_level']].copy()
test_y = test_df['price'].copy()

poly = sklearn.preprocessing.PolynomialFeatures(degree=10)
train_x = poly.fit_transform(train_x_raw)
test_x = poly.fit_transform(test_x_raw)

regression_model = LinearRegression()
regression_model.fit(train_x, train_y)
prediction_train = regression_model.predict(train_x)
prediction_test = regression_model.predict(test_x)

regression_model_no_poly = LinearRegression()
regression_model_no_poly.fit(train_x_raw, train_y)
prediction_train_no_poly = regression_model_no_poly.predict(train_x_raw)
prediction_test_no_poly = regression_model_no_poly.predict(test_x_raw)

train_df['regression_prediction'] = prediction_train
test_df['regression_prediction'] = prediction_test

regression_train_mae = mean_absolute_error(train_y, prediction_train)
regression_test_mae = mean_absolute_error(list(dataframe['price']), prediction_test)
result_MAE.loc[len(result_MAE)] = ['linear_regression', f'{regression_train_mae:.6f}', f'{regression_test_mae:.6f}']

regression_train_rmse = root_mean_squared_error(train_y, prediction_train)
regression_test_rmse = root_mean_squared_error(list(dataframe['price']), prediction_test)
result_RMSE.loc[len(result_RMSE)] = ['linear_regression', f'{regression_train_rmse:.6f}', f'{regression_test_rmse:.6f}']

regression_r2_train = r2_score(train_y, prediction_train)
regression_r2_test = r2_score(list(dataframe['price']), prediction_test)
result_R2.loc[len(result_R2)] = ['linear_regression', f'{regression_r2_train:.6f}', f'{regression_r2_test:.6f}']

regression_no_poly_train_mae = mean_absolute_error(train_y, prediction_train_no_poly)
regression_no_poly_test_mae = mean_absolute_error(list(dataframe['price']), prediction_test_no_poly)
result_MAE.loc[len(result_MAE)] = ['linear_regression_no_poly', f'{regression_no_poly_train_mae:.6f}', f'{regression_no_poly_test_mae:.6f}']

regression_no_poly_train_rmse = root_mean_squared_error(train_y, prediction_train_no_poly)
regression_no_poly_test_rmse = root_mean_squared_error(test_y, prediction_test_no_poly)
result_RMSE.loc[len(result_RMSE)] = ['linear_regression_no_poly', f'{regression_no_poly_train_rmse:.6f}', f'{regression_no_poly_test_rmse:.6f}']

regression_no_poly_r2_train = r2_score(train_y, prediction_train_no_poly)
regression_no_poly_r2_test = r2_score(test_y, prediction_test_no_poly)
result_R2.loc[len(result_R2)] = ['linear_regression_no_poly', f'{regression_no_poly_r2_train:.6f}', f'{regression_no_poly_r2_test:.6f}']

print(f'result_MAE\n{result_MAE}\n')
print(f'result_RMSE\n{result_RMSE}')
```
